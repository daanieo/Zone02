Data analysis for GAMA's zone02 output

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("result.txt",sep=",",skiprows=1,header=None)
data = data.drop(columns=[0,3,1443])
data.head(10)
    
```

```python
for i in range(len(data)):
    plt.plot(data.iloc[i,:])

plt.title("Filling up the model")
plt.ylabel("hungry days")
plt.xlabel("ticks")
plt.ylim([0,8500])
plt.xlim([0,1450])
```
